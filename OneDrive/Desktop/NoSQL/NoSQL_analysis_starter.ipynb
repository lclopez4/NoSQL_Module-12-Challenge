### 1. Which establishments have a hygiene score equal to 20?

```python
from pprint import pprint

# Assign the establishments collection to a variable
establishments = db['establishments']

# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}
result = list(establishments.find(query))

# Use count_documents to display the number of documents in the result
print(establishments.count_documents({'scores.Hygiene': {'$eq': 20}}))
# Display the first document in the results using pprint
if result:
	pprint(result[0])
else:
	print("No documents found with a hygiene score of 20.")
```

```python
import pandas as pd

# Convert the result to a Pandas DataFrame
hygiene_20_df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
print(f"Number of rows: {len(hygiene_20_df)}")
# Display the first 10 rows of the DataFrame
hygiene_20_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
#find penang flavor's lat and lon
list(establishments.find({'BusinessName':'Penang Flavours'}))
#query for question 3
#list the top 5 restaurants near penang flavours within the degree search parameters, sorted by hygiene score ascending
degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000
query = {'RatingValue': {'$eq': 5},
         'geocode.latitude':{'$gte': (latitude - degree_search), '$lte': (latitude + degree_search)},
         'geocode.longitude':{'$gte': (longitude - degree_search), '$lte': (longitude + degree_search)}
         }
sort =  [('scores.Hygiene', 1)] # for sorting, 1 is ascending, -1 is descending
limit = 5
results = list(establishments.find(query).sort(sort).limit(limit))
pprint(results)
```

```python
# Convert result to Pandas DataFrame
near_penang_df = pd.DataFrame(results)
near_penang_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# , 'LocalAuthorityName': {'$regex': "London"}
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}
group_query = {'$group': {'_id': {'LocalAuthorityName': '$LocalAuthorityName'},
                          'count':{'$sum':1}
                          }}
sort_values = {'$sort': {'count': -1}}

pipeline = [match_query, group_query, sort_values]
results = list(establishments.aggregate(pipeline))
results
# Print the number of documents in the result
print(f"Number of documents in result: {establishments.count_documents(query)}")
# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_0_by_authority_df = pd.DataFrame(results)
hygiene_0_by_authority_df
# Display the number of rows in the DataFrame
len(hygiene_0_by_authority_df)
# Display the first 10 rows of the DataFrame
hygiene_0_by_authority_df.head(10)
```
